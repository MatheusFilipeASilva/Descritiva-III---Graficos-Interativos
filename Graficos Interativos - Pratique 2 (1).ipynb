#### 1. Carregue os dados da tarefa anterior

````python
from  pandas_datareader import data as web

import plotly.express as px
from plotly import graph_objects
from plotly.graph_objs import Layout

import pandas as pd
import numpy as np
import yfinance as yf
````

````python
ativos = ['^BVSP', 'BTC-USD', 'BRL=X']

dt_srt, dt_end = '2020-01-01', '2024-09-16'

yframe = yf.download(ativos, 
                    start = dt_srt, 
                    end = dt_end)
yframe
````

#### 2. Média móvel

1. Faça uma média móvel  em 90 dias dos valores de fechamento ajustados dos ativos que você escolheu
2. Utilizando a fução ```rolling()```, calcule o desvio pardão desses ativos ao longo do tempo, utilizando a mesma janela móvel do item acima
2. Monte um gráfico de linhas para as médias móveis dos ativos da sua base
4. Monte um gráfico de linhas para os desvios padrão em janela móvel
5. Compare os dois gráficos que você fez com o da tarefa anterior. Veja se eles apontam conclusões semelhantes quanto aos períodos de maior volatilidade (variação) dos ativos.

````python
y_close = yframe['Close']
y_close.columns = ['BRL=X', 'BTC-USD', '^BVSP']
date = pd.date_range('2020-01-01', '2024-09-15')

y_close.index = date
y_close = y_close.ffill(axis = 1)

mm = y_close.rolling(3).mean()
desvio_padrao = y_close.rolling(3).std()
mm


# por algum motivo, o BRL=X não está aplicando corretamente o ffill... não consegui identificar o porque...
````

````python

mm = mm.stack().reset_index()
mm.columns = ['data', 'ativo', 'valor']

desvio_padrao = desvio_padrao.stack().reset_index()
desvio_padrao.columns = ['data', 'ativo', 'valor']

mm
````

````python
# Gráfico 1:
fig1 = px.line(mm, x = 'data', y='valor', color='ativo')
fig1.show()
````

````python
fig2 = px.line(desvio_padrao, x='data', y='valor', color='ativo')
fig2.show()
````

#### 3. Gráfico de *candlestick*
1. Selecione o período dos últimos 60 dias (corridos) para um dos ativos da sua base
2. Monte um gráfico de *candlestick* para esses dados.

````python

lista_indicadores = [row[0] for row in yframe.columns[::3]]
lista_indicadores
````

````python
ativos = ['BTC-USD']*len(lista_indicadores)
colunas = list(zip(lista_indicadores, ativos))
colunas
````

````python
btc_60 = yframe.loc['2024-07-16':,colunas]
# btc_60.index = btc_60['Date']
# btc_60 = btc_60.drop(level = 1, labels='Ticker')
btc_60.columns = btc_60.columns.droplevel(1)
btc_60
````

````python
graph = {'x': btc_60.index, 
        'open': btc_60.Open, 
        'close': btc_60.Close, 
        'high': btc_60.High, 
        'low': btc_60.Low, 
        'type': 'candlestick', 
        'name': 'BTC-USD',  
        'showlegend': True}
````

````python
layout = Layout(
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)'
)

grafico = graph_objects.Figure(
    data=[graph],
    layout_title='BTC vs USD',
    layout = layout
)
````

````python
grafico.update_xaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')
grafico.update_yaxes(showgrid=True, gridwidth=1, gridcolor='LightGrey')

grafico
````
